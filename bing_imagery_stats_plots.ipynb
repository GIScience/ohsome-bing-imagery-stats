```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('imagery_stats_per_month.csv')
df["month"] = pd.to_datetime(df["month"])
df.sort_values(by="month", inplace=True)
df.drop(df.tail(1).index,inplace=True) 
```

```python
pd.set_option('display.max_rows', 500)
columns = [
    "month",
    "n_edits_million",
    "n_users",
    "bing_n_edits_million",
    "bing_n_users"
]

display(df[columns])
```

## Absolute and relative number of monthly map edits in OpenStreetMap

```python
fig, ax1 = plt.subplots(figsize = (12, 8))

ax1.bar(
    df["month"],
    df["n_edits_million"],
    width=20,
    label="Overall",
    color="lightgrey"
)

ax1.bar(
    df["month"],
    df["bing_n_edits_million"],
    width=20,
    label='Bing Imagery',
    color="blue",
)

ax1.set_ylim((0, 40))
ax1.set_xlim(
    datetime.datetime(2019, 12, 1),
    datetime.datetime(2025, 6, 1),
)

ax1.set_ylabel("Edits [Million]", color="blue", fontsize=14)
ax1.tick_params(axis="y", labelcolor="blue")
ax1.legend(loc="upper left")

ax2 = ax1.twinx()  

ax2.plot(
    df["month"],
    100 * df["bing_n_edits_million"] / df["n_edits_million"],
    color="red",
    linewidth=2,
    label="Bing Imagery"
)

ax2.set_ylim(0, 100)
ax2.set_ylabel("Share [%]", color="red", fontsize=14)
ax2.tick_params(axis="y", labelcolor="red")

ax2.legend(loc="upper right")

plt.legend()
plt.title('Absolute and relative number of monthly map edits in OpenStreetMap using Bing Imagery')

plt.savefig('monthly_map_edits.png', dpi=300)
plt.show()
```

## Absolute and relative number of monthly contributors in OpenStreetMap

```python
fig, ax1 = plt.subplots(figsize = (12, 8))

ax1.bar(
    df["month"],
    df["n_users"],
    width=20,
    label="Overall",
    color="lightgrey"
)

ax1.bar(
    df["month"],
    df["bing_n_users"],
    width=20,
    label='Bing Imagery',
    color="blue",
)

ax1.set_ylim((0, 70*10**3))
ax1.set_xlim(
    datetime.datetime(2019, 12, 1),
    datetime.datetime(2025, 6, 1),
)

ax1.set_ylabel("Contributors", color="blue", fontsize=14)
ax1.tick_params(axis="y", labelcolor="blue")
ax1.legend(loc="upper left")

ax2 = ax1.twinx()  

ax2.plot(
    df["month"],
    100 * df["bing_n_users"] / df["n_users"],
    color="red",
    linewidth=2,
    label="Bing Imagery"
)

ax2.set_ylim(0, 100)
ax2.set_ylabel("Share [%]", color="red", fontsize=14)
ax2.tick_params(axis="y", labelcolor="red")

ax2.legend(loc="upper right")

plt.legend()
plt.title('Absolute and relative number of monthly contributors in OpenStreetMap using Bing Imagery')


plt.savefig('monthly_contributors.png', dpi=300)
plt.show()
```

## Editor Software

```python
df2 = pd.read_csv('imagery_stats_per_editor.csv')
df2.sort_values(by="n_edits_million", ascending=True, inplace=True)
```

display(df2)

## Absolute number of map edits in OpenStreetMap using Bing Imagery per Editor

```python
fig, ax1 = plt.subplots(figsize = (12, 4))


hbars1 = ax1.barh(
    df2["editor"],
    df2["n_edits_million"],
    color="lightgrey",
    label="Overall"
)


hbars2 = ax1.barh(
    df2["editor"],
    df2["bing_n_edits_million"],
    color="blue",
    label="Bing Imagery"
)

ax1.bar_label(
    hbars2,
    fmt='%.2f',
    label_type='edge',
    padding=5,
    color="blue"
)

ax1.legend(loc="lower right")
ax1.set_xlabel("Edits [Million]")
ax1.set_xlim(0, 850)

plt.title("Absolute number of map edits in OpenStreetMap using Bing Imagery per Editor")

plt.savefig('edits_editor.png', dpi=300)
plt.show()
```

## Absolute number of contributors in OpenStreetMap using Bing Imagery per Editor

```python
fig, ax1 = plt.subplots(figsize = (12, 4))


hbars1 = ax1.barh(
    df2["editor"],
    df2["n_users"] / 1000,
    color="lightgrey",
    label="Overall"
)


hbars2 = ax1.barh(
    df2["editor"],
    df2["bing_n_users"] / 1000,
    color="blue",
    label="Bing Imagery"
)

ax1.bar_label(
    hbars2,
    fmt="%.1f",
    label_type='edge',
    padding=5,
    color="blue"
)

ax1.legend(loc="lower right")
ax1.set_xlabel("Contributors [Thousands]")
ax1.set_xlim(0, 1000)

plt.title("Absolute number of contributors in OpenStreetMap using Bing Imagery per Editor")

plt.savefig('contributors_editor.png', dpi=300)
plt.show()
```
